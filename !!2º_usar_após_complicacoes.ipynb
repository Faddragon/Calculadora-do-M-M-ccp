#### 💡Preencher a coluna "TEMPO_INTERNACAO_HORAS" e "TEMPO_INTERNACAO_DIAS"

```python
# Garantir que as colunas estejam no formato datetime
df_raw['INICIO'] = pd.to_datetime(df_raw['INICIO'], errors='coerce')
df_raw['HORA_ALTA'] = pd.to_datetime(df_raw['HORA_ALTA'], errors='coerce')

# Calcular a diferença em horas e dias
df_raw['TEMPO_INTERNACAO_HORAS'] = (df_raw['HORA_ALTA'] - df_raw['INICIO']).dt.total_seconds() / 3600
df_raw['TEMPO_INTERNACAO_DIAS'] = df_raw['TEMPO_INTERNACAO_HORAS'] / 24
```

#### *️⃣ Contagem por cirurgião

```python
# Contagem dos grupos
grupo_counts = df_raw['CIRURGIA_GRUPO'].value_counts().reset_index()
grupo_counts.columns = ['CIRURGIA_GRUPO', 'Quantidade']

# Calcular altura ideal: 40px por grupo
altura = max(400, len(grupo_counts) * 40)

# Gráfico de barras horizontais
fig = px.bar(
    grupo_counts,
    x='Quantidade',
    y='CIRURGIA_GRUPO',
    orientation='h',
    color='CIRURGIA_GRUPO',
    title='Contagem de procedimentos - Janeiro-junho 2025',
    text='Quantidade',
    color_discrete_sequence=px.colors.qualitative.Safe
)

fig.update_layout(
    height=altura,
    xaxis_title='Número de Casos',
    yaxis_title='Grupo de Cirurgia',
    template='simple_white',
    yaxis={'categoryorder': 'total ascending'},
    showlegend=False  # Removido pois eixo Y já representa os grupos
)

fig.show()

```
